STEP 2

```python
train.isnull().sum()
```

Some of the variables have missing data, particularly those that pertain to the garages, basements, and the lot. These values which are missing must be addressed prior to analysis.

```python
train["Alley"] = train["Alley"].fillna("None")
train["BsmtQual"] = train["BsmtQual"].fillna("None")
train["FireplaceQu"] = train["FireplaceQu"].fillna("None")
train["GarageType"] = train["GarageType"].fillna("None")
```

```python
train["LotFrontage"] = train["LotFrontage"].fillna(train["LotFrontage"].mean())
train["GarageYrBlt"] = train["GarageYrBlt"].fillna(train["GarageYrBlt"].mean())
```

```python
train["GarageArea"] = train["GarageArea"].fillna(0)
train["TotalBsmtSF"] = train["TotalBsmtSF"].fillna(0)
```

```python
train.isnull().sum()
```

Dealt with missing values by filling categorical variables with the value of None and numerical variables with a basic value like the mean or zero. This will make sure that the dataset is not filled with missing values and is prepared to be analyzed further.

```python
plt.boxplot(train["SalePrice"])
plt.title("Sale Price After Cleaning")
plt.show()
```

The SalePrice variable still has outliers, of which the bulk are on the high end. These values were preserved as they are costly houses which could be significant in prediction.

```python
train["Id"].nunique()
len(train)
```

```python
train = train.drop(columns=["Id"])
```

```python
train.columns
```

Id column was dropped since it is merely an identifier for each house and that does not give any valuable information about the house. The values in the Id column are unique and, therefore, do not aid in explaining and predicting SalePrice. Retaining it in the dataset may introduce noise to the analysis so it was not subjected to any further steps.

```python
train.shape
```

After cleaning, the data format was also a little different since the Id column was removed but the number of observations did not change.

PART 3

```python
numeric_data = train.select_dtypes(include=["int64", "float64"])
```

```python
correlation_matrix = numeric_data.corr()
correlation_matrix
```

This is a correlation matrix indicating links among numeric variables. A higher value of 1 or -1 denotes stronger relationships whereas a close value of 0 denotes weak relationships.

```python
features = [
    "SalePrice",
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "1stFlrSF",
    "FullBath",
    "TotRmsAbvGrd",
    "YearBuilt"
]

selected_corr = train[features].corr()

plt.figure(figsize=(10, 8))
plt.imshow(selected_corr)
plt.colorbar()
plt.title("Correlation Heatmap (Selected Features)")
plt.xticks(range(len(features)), features, rotation=90)
plt.yticks(range(len(features)), features)
plt.show()
```

This heatmap concentrates on some key features which are important and collinearity becomes more visible. Certain size related features are correlated with one another, i.e. they can share a point of information.

```python
correlation_matrix["SalePrice"].sort_values(ascending=False)
```

Through correlation analysis, it is found that there are numeric features that have moderate positive correlation with SalePrice. FullBath, Total Rooms Above Ground, Year Built and Year Remodeled feature are more correlated which implies that houses with more bathrooms, rooms and recent construction are more likely to be sold at a higher price.

For feature selection, the Id column was previously excluded due to the fact that it is nothing but a feature is only a feature identifier and it does not assist in predicting SalePrice. In Part 3, there is some collinearity in its features. None of the other features were however eliminated at this point. The future modeling will take these relationships into account.

This project covered the housing dataset in terms of visualizations and correlation analysis to learn the factors associated with SalePrice. The findings revealed a strong relationship between price and house size and the overall quality and that a few variables are skewed to the right and have outliers.
